```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re


link='https://www.tutiempo.net/clima/ws-766120.html'
link_gen = 'https://www.tutiempo.net'
info = requests.get(link)
soup = BeautifulSoup(info.content, 'html5lib')
urls = [[link_gen+i['href'],i['title'].replace('Clima en GUADALAJARA en el año ','')]
        for i in soup.find_all('a') if 'Clima ' in i.text]
year_data=[]

#Loop to obtain the data for the years and months
for j, link_page in enumerate(urls):    
    #Generate a variable called classes with the information of each year
    #This variable contains all the months
    classes = BeautifulSoup(requests.get(urls[j][0]).content, 'html5lib')
    
    #Each year is broken down into months, they are stored in the 'a' web identifier
    month_url = [link_gen+hist['href'] for hist in classes.find_all('a') if 'title' in hist.attrs 
                 and 'Datos históricos climáticos en GUADALAJARA' in hist['title']]
    month = [re.sub(r'[0-9]','',hist['title'].replace('Datos históricos climáticos en GUADALAJARA (','').replace(')','').replace(' ',''))
            for hist in classes.find_all('a') if 'title' in hist.attrs 
            and 'Datos históricos climáticos en GUADALAJARA' in hist['title']]
    
    monthly_data = []
    # Looking for the tables that appear in each month
    for i,mon in enumerate(month_url):
        table = BeautifulSoup(requests.get(month_url[i]).content, 'html5lib')
        
        #Tables have the tbody parameter, in the position 3.
        # they can be broken down into 'tr' parameters
        d=table.find_all('tbody')[3].find_all('tr')[:-2]
        s=[]
        for element in d:
            info_list = element.find_all('td')
            s.append([inf.text for inf in info_list][1:])
            
        df = pd.DataFrame(s)
        df['day']=np.arange(0,df.shape[0])
        df['month']=month[i]
        df['year']= int(urls[j][1])*np.ones(df.shape[0])

        monthly_data.append(df)
    year_data.append(pd.concat(monthly_data))
    
all_data=pd.concat(year_data,ignore_index=True)
all_data.columns = ['T', 'Tm', 'Tm', 'SLP', 'H', 'PP',
                    'VV', 'V', 'VM', 'VG', 'RA', 'SN', 
                    'TS', 'FG', 'day', 'month','year']
all_data[(all_data.year==2019) & (all_data.month=='Enero')]
```
